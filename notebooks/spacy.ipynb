```python
import spacy
from spacy.matcher import Matcher

# Türkçe transformer modelini yükleyin
nlp = spacy.load("tr_core_news_trf")

# Örnek metin (Özel Eğitim ile ilgili makalelerden alınmış gibi)
text = """
Özel gereksinimli bireylerin eğitime erişimi büyük önem taşımaktadır.
Özel eğitime ihtiyacı olan çocuklar, uygun eğitim ortamlarında bulunmalıdır.
Geleneksel eğitim sistemleri, özel eğitime muhtaç bireyler için yeterli olmayabilir.
Özel eğitim gerektiren öğrencilerin hakları anayasal güvence altındadır.
Bazı okullar, özel eğitim programlarıyla destek sağlıyor.
"""

doc = nlp(text)

# Yöntem 1: İsim öbeklerinden ("noun chunks") "özel" içeren terimleri seçelim.
"""noun_chunks_with_ozel = [chunk.text for chunk in doc.noun_chunks if "özel" in chunk.text.lower()]
print("Terimler:")
for i in noun_chunks_with_ozel:
    print(i)
#print(noun_chunks_with_ozel)
"""
# Yöntem 2: Matcher kullanarak "özel" ile başlayan kalıpları bulalım.
matcher = Matcher(nlp.vocab)
# Bu kalıp, "özel" kelimesi ile başlayıp, ardından sıfır veya daha fazla herhangi bir token ve sonrasında en az bir alfabetik token gelecek şekilde ayarlanmıştır.
pattern = [{"LOWER": "özel"}, {"OP": "*"}, {"IS_ALPHA": True, "OP": "+"}]
matcher.add("SPECIAL_EDU", [pattern])
matches = matcher(doc)
matched_terms = set()
for match_id, start, end in matches:
    span = doc[start:end]
    matched_terms.add(span.text)
print("Matcher Yöntemi ile Bulunan Terimler:")
#print(type(matched_terms))
for i in matched_terms:
    print(i)
#print(list(matched_terms))
```

```python
import spacy
from collections import Counter

# Modelinizi yükleyin (örneğin, transformer tabanlı model)
nlp = spacy.load("tr_core_news_trf")

text = """
Özel gereksinimli bireylerin eğitime erişimi büyük önem taşımaktadır.
Özel eğitime ihtiyacı olan çocukların uygun eğitim ortamlarında bulunması gerekmektedir.
Geleneksel eğitim sistemleri, özel eğitime muhtaç bireyler için yeterli olmayabilir.
Bazı öğrenciler, özel eğitim gerektiren derslere ihtiyaç duymaktadır.
Destek eğitimi alan öğrenciler başarılarını artırmaktadır.
"""

doc = nlp(text)

# İsim öbeklerini çıkarın ve küçük harfe çevirin
noun_chunks = [chunk.text.lower() for chunk in doc.noun_chunks]

# Çok kısa öbekleri (örneğin tek kelimelik veya 1-2 karakterli) filtreleyebilirsiniz
filtered_chunks = [chunk for chunk in noun_chunks if len(chunk) > 3]

# Frekansları sayın
chunk_freq = Counter(filtered_chunks)

print("En sık geçen isim öbekleri:")
for chunk, freq in chunk_freq.most_common(10):
    print(f"{chunk}: {freq}")
```

```python
import spacy

# Örnek metin
with open("data/contractions/sayi_11.txt", "r", encoding="utf-8") as file:
    text = file.read()

custom_stopwords = set([
    "ile", "ve", "bir", "bu", "şu", "çok", "gibi", "ama", "daha", "olan", "önemli", "için", "ise"
])

tokens = [word for word in text.split() if word not in custom_stopwords]

nlp = spacy.load("tr_core_news_trf")

tokens = [token.lemma_ for token in nlp(" ".join(tokens))]
print(tokens)
```
